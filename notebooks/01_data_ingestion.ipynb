# Step 6: Explore Fund Master

In this section, we load the `01_fund_master.csv` dataset to understand its structure, print the required unique values, perform some additional deep-dive EDA, and analyze the AMFI scheme code structure.

```python
import pandas as pd
import os

# Determine correct path to data
data_path = os.path.join('..', 'data', 'raw', '01_fund_master.csv')
fund_master = pd.read_csv(data_path)

print("--- Dataset Overview ---")
display(fund_master.head(3))
```

### 1. Required Unique Values Exploration

```python
print(f"Unique Fund Houses ({fund_master['fund_house'].nunique()}):\n{list(fund_master['fund_house'].unique())}\n")
print(f"Unique Categories ({fund_master['category'].nunique()}):\n{list(fund_master['category'].unique())}\n")
print(f"Unique Sub-Categories ({fund_master['sub_category'].nunique()}):\n{list(fund_master['sub_category'].unique())}\n")
print(f"Unique Risk Grades ({fund_master['risk_category'].nunique()}):\n{list(fund_master['risk_category'].unique())}\n")
```

### 2. Additional Deep-Dive EDA 
Going beyond the basics to truly understand the dataset composition.

```python
print("--- Number of Schemes per Fund House ---")
display(fund_master['fund_house'].value_counts())

print("\n--- Average Expense Ratio by Category ---")
display(fund_master.groupby('category')['expense_ratio_pct'].mean().round(2))

print("\n--- Number of Schemes per Risk Grade ---")
display(fund_master['risk_category'].value_counts())
```

### 3. AMFI Scheme Code Structure
The AMFI code (`amfi_code`) is a unique 6-digit integer assigned by the Association of Mutual Funds in India (AMFI). 
- It serves as the **Primary Key** for any mutual fund database in India.
- It uniquely identifies not just the scheme, but the specific **plan** (e.g., Regular vs Direct) and **option** (e.g., Growth vs IDCW). This is why a single fund house can have hundreds of AMFI codes.
- In our dataset, we can see that codes like `125497` (HDFC Top 100) are mapped exactly to a specific AMC, Category, and Risk Grade. 
- Because it is an industry-standard identifier, it is universally used across platforms (like NSE, BSE, Kuvera, Zerodha) and public APIs (like `mfapi.in/mf/{amfi_code}`) to fetch historical data accurately without relying on string-matching fund names.
